```python
#Q4, part B

import matplotlib.pyplot as plt
import numpy as np

# data from results table
t = np.array([3.4, 6.85, 10.3, 13.75, 17.2, 20.7, 24.15, 27.7, 31.05, 34.55])  # time in seconds
ln_peak = np.array([1.196948189, 0.798857635, 0.401457087, 0.003992021, -0.394525168,
                    -0.796287939, -1.197328262, -1.62455155, -2.024953356, -2.453407983])  # natural log of Peak Value

# plotting
plt.figure(figsize=(10, 6))
plt.plot(t, ln_peak, 'o-', color='blue')
plt.title('Natural Log of Peak Amplitude vs Time')
plt.xlabel('Time (s)')
plt.ylabel('Ln(Peak Amplitude)')
plt.grid(True)
plt.show()

##################################################################################################################################
#Q4, part C ...underdamped harmonic oscillator


import numpy as np
import matplotlib.pyplot as plt

# data
t = np.array([3.4, 6.85, 10.3, 13.75, 17.2, 20.7, 24.15, 27.7, 31.05, 34.55])  # Time in seconds
amplitude = np.array([3.31, 2.223, 1.494, 1.004, 0.674, 0.451, 0.302, 0.197, 0.132, 0.086])  # Amplitude of oscillations

# we are assuming the first value corresponds to the maximum amplitude and subsequent values are the peaks
# we will normalize the amplitude by the first value to represent the decay relative to the initial amplitude
normalized_amplitude = amplitude / amplitude[0]

# plotting the normalized amplitude over time
plt.figure(figsize=(10, 6))
plt.plot(t, normalized_amplitude, 'o-', color='blue', label='Underdamped Oscillation')

# fitting an exponential decay to the data
# we'll use the natural logarithm of the amplitude for the fitting
ln_amplitude = np.log(amplitude)
coefficients = np.polyfit(t, ln_amplitude, 1)  # linear fit since ln(amplitude) should decay linearly
decay_fit = np.poly1d(coefficients)

# plotting the exponential decay fit
t_fit = np.linspace(min(t), max(t), 100)
amplitude_fit = np.exp(decay_fit(t_fit))
plt.plot(t_fit, amplitude_fit, 'r--', label='Exponential Decay Fit')

# labels and title
plt.title('Underdamped Oscillation and Exponential Decay Fit')
plt.xlabel('Time (s)')
plt.ylabel('Normalized Amplitude')
plt.legend()
plt.grid(True)
plt.show()

######################################################################################################################################################
import numpy as np
import matplotlib.pyplot as plt

# data
t = np.array([3.4, 6.85, 10.3, 13.75, 17.2, 20.7, 24.15, 27.7, 31.05, 34.55])  # Time in seconds
peak_values = np.array([3.31, 2.223, 1.494, 1.004, 0.674, 0.451, 0.302, 0.197, 0.132, 0.086])  # Peak Values

# plotting the peak values for the critically damped case
plt.figure(figsize=(10, 6))
plt.plot(t, peak_values, 'o-', color='darkorange', label='Critically Damped Peaks')

# labels and title
plt.title('Critically Damped Oscillation Peaks Over Time')
plt.xlabel('Time (s)')
plt.ylabel('Peak Value')

# plotting
plt.grid(True)
plt.legend()
plt.show()

```
